# Confidence Intervals

```python
def seed_confidence(
    val_list: list, 
    confidence: float = 0.95, 
    str_dp: int = 3,
    t_val: bool = False):
    # get confidence interval for n values representing different seeds

    n_seeds = len(val_list)
    
    if t_val:
        # get tval for given confidence, ie. treat examples as samples from larger distribution
        modifier = scistat.t.ppf((1+confidence)/2.0, df=n_seeds-1)
    else:
        # get zval, ie treat examples as distribution
        modifier = scistat.norm.ppf((1+confidence)/2.0)

    # calculate standard error as standard deviation over squareroot of number of examples
    stdev = np.std(val_list, ddof=1)
    stderr = stdev/np.sqrt(n_seeds)
    
    confint = modifier * stderr
    
    confint_upper = np.mean(val_list) + confint
    confint_lower = np.mean(val_list) - confint

    return confint, confint_upper, confint_lower, f"{np.mean(val_list):.{str_dp}f} ± {confint:.{str_dp}f}"
```

```python
def make_f1_confidence_dict(
    result_df: pd.DataFrame, 
    technique: str, 
    batch_sizes = [0,16,32,64,128,256,512,1024], 
    exp_date: dt.datetime.date = None, 
    balanced: bool = None,
    prompt: str = None, 
    model:str = None):

    df = result_df[result_df.technique==technique].copy()
    df['datetime'] = pd.to_datetime(df['datetime'])

    if exp_date is not None: df = df[df.datetime.dt.date==exp_date]
    if balanced is not None: df = df[df.balanced_train==balanced]
    if prompt is not None: df = df[df.template==prompt]
    if model is not None: df = df[df.model==model]

    dict_confidence_attr = {
        'technique':technique,
        'model':model,
        'date': exp_date,
        'balanced_train': balanced,
        'prompt': prompt
    }

    dict_confidence_batches = {b:None for b in batch_sizes}

    for batchsize, bdf in df.groupby('n_train'):
        if bdf.shape[0] != bdf.seed.nunique(): 
            error_dict = dict_confidence_attr
            error_dict['batchsize'] = batchsize
            raise Exception(f'Duplicate seed vales  - {bdf.shape[0]} rows vs {bdf.seed.nunique()} seeds / {error_dict}')
        else: dict_confidence_batches[batchsize] = seed_confidence(bdf.f1.tolist())[3]
    
    dict_confidence_attr.update(dict_confidence_batches)
    
    return dict_confidence_attr
            
```

```python
def seed_confidence(
    val_list: list, 
    confidence: float = 0.95, 
    str_dp: int = 2,
    t_val: bool = False):
    # get confidence interval for n values representing different seeds

    n_seeds = len(val_list)
    
    if t_val:
        # get tval for given confidence, ie. treat examples as samples from larger distribution
        modifier = scistat.t.ppf((1+confidence)/2.0, df=n_seeds-1)
    else:
        # get zval, ie treat examples as distribution
        modifier = scistat.norm.ppf((1+confidence)/2.0)

    # calculate standard error as standard deviation over squareroot of number of examples
    stdev = np.std(val_list, ddof=1)
    stderr = stdev/np.sqrt(n_seeds)
    
    confint = modifier * stderr
    
    confint_upper = np.mean(val_list) + confint
    confint_lower = np.mean(val_list) - confint

    return confint, confint_upper, confint_lower, f"{np.mean(val_list):.{str_dp}f} ± {confint:.{str_dp}f}"

def make_f1_confidence_dict(
    result_df: pd.DataFrame, 
    technique: str, 
    batch_sizes = [0,16,32,64,128,256,512,1024], 
    exp_date: dt.datetime.date = None, 
    balanced: bool = None,
    prompt: str = None, 
    model:str = None):

    df = result_df[result_df.technique==technique].copy()
    df['datetime'] = pd.to_datetime(df['datetime'])

    if exp_date is not None: df = df[df.datetime.dt.date==exp_date]
    if balanced is not None: df = df[df.balanced_train==balanced]
    if prompt is not None: df = df[df.template==prompt]
    if model is not None: df = df[df.model==model]

    dict_confidence_attr = {
        'Technique':technique.replace('_',' ').title(),
        'Model':model,
        'Date': exp_date,
        'Balanced Train': balanced,
        'Prompt': prompt.replace("{\"placeholder\":\"text_a\"}", "[text]").replace("{\"mask\"}", "") if prompt is not None else prompt
    }

    dict_confidence_batches = {b:None for b in batch_sizes}

    for batchsize, bdf in df.groupby('n_train'):
        if bdf.shape[0] != bdf.seed.nunique(): 
            error_dict = dict_confidence_attr
            error_dict['batchsize'] = batchsize
            raise Exception(f'Duplicate seed values  - {bdf.shape[0]} rows vs {bdf.seed.nunique()} seeds / {error_dict}')
        else: dict_confidence_batches[batchsize] = seed_confidence(bdf.f1.tolist())[3]
    
    dict_confidence_attr.update(dict_confidence_batches)
    
    return dict_confidence_attr
```
